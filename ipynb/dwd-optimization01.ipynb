```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_aspect(1)
ax.grid()

circle = plt.Circle((0, 0), 1, color='blue', fill=False)
ax.add_patch(circle)

ax.plot([-1-np.sqrt(2), 1], [-1, 1+np.sqrt(2)], color='red', linewidth='2', zorder=-0.1)
ax.plot([-1.5, 1.0], [-1.0, 1.5], color='#ffcccc', zorder=-0.1)
ax.plot([-1, 1.5], [-1-np.sqrt(2), 1.5-np.sqrt(2)], color='#ffdddd', zorder=-0.1)

circle = plt.Circle((-np.sqrt(2)/2, np.sqrt(2)/2), .03, color='red')
ax.add_patch(circle)
circle = plt.Circle((-.25 + np.sqrt(7)/4, .25 + np.sqrt(7)/4), .03, color='#ffcccc')
ax.add_patch(circle)
circle = plt.Circle((-.25 - np.sqrt(7)/4, .25 - np.sqrt(7)/4), .03, color='#ffcccc')
ax.add_patch(circle)
circle = plt.Circle((np.sqrt(2)/2, -np.sqrt(2)/2), .03, color='#ffdddd')
ax.add_patch(circle)

circle = plt.Circle((0, np.sqrt(2)), .05, color='blue')
ax.add_patch(circle)
circle = plt.Circle((0, .5), .05, color='#ccccff')
ax.add_patch(circle)
circle = plt.Circle((0, -np.sqrt(2)), .05, color='#ddddff')
ax.add_patch(circle)

ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)

plt.show()
```
