```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_aspect(1)
ax.grid()

circle = plt.Circle((2, 1), .05, color='blue', zorder=5)
ax.add_patch(circle)
ax.annotate(r'$\xi_1 = (2, 1)$', xy=(2+.2, 1), fontsize=10)
ax.annotate(r'$t_1 = +1$', xy=(2+.2, 1-.2), fontsize=10)

circle = plt.Circle((3, -1), .05, color='blue', zorder=5)
ax.add_patch(circle)
ax.annotate(r'$\xi_2 = (3, -1)$', xy=(3-1, -1), fontsize=10)
ax.annotate(r'$t_2 = -1$', xy=(3-1, -1-.2), fontsize=10)

circle = plt.Circle((0, 0), .05, color='blue', zorder=5)
ax.add_patch(circle)
ax.annotate(r'$\xi_3 = (0, 0)$', xy=(0+.2, 0), fontsize=10)
ax.annotate(r'$t_3 = +1$', xy=(0+.2, 0-.2), fontsize=10)

ax.plot([-.5, 3.5], [-1.5, .5], color='red', linewidth='3')

ax.plot([2, 2.5], [1, 0], color='green', linestyle = 'dashed')
ax.plot([2.5, 3], [0, -1], color='purple', linestyle = 'dashed')
ax.plot([0, .5], [0, -1], color='orange', linestyle = 'dashed')


ax.set_xlim(-0.5, 3.5)
ax.set_ylim(-1.5, 1.5)

plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_aspect(1)
ax.grid()

circle = plt.Circle((2, 1), .05, color='blue', zorder=5)
ax.add_patch(circle)
ax.annotate(r'$\xi_1 = (2, 1)$', xy=(2+.2, 1), fontsize=10)
ax.annotate(r'$t_1 = +1$', xy=(2+.2, 1-.2), fontsize=10)

circle = plt.Circle((3, -1), .05, color='blue', zorder=5)
ax.add_patch(circle)
ax.annotate(r'$\xi_2 = (3, -1)$', xy=(3-1, -1), fontsize=10)
ax.annotate(r'$t_2 = -1$', xy=(3-1, -1-.2), fontsize=10)

circle = plt.Circle((0, 1), .05, color='blue', zorder=5)
ax.add_patch(circle)
ax.annotate(r'$\xi_3 = (0, 1)$', xy=(0+.2, 1), fontsize=10)
ax.annotate(r'$t_3 = +1$', xy=(0+.2, 1-.2), fontsize=10)

ax.plot([-.5, 3.5], [-1.5, .5], color='red', linewidth='3')

ax.plot([2, 2.5], [1, 0], color='green', linestyle = 'dashed')
ax.plot([2.5, 3], [0, -1], color='purple', linestyle = 'dashed')


ax.set_xlim(-0.5, 3.5)
ax.set_ylim(-1.5, 1.5)

plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_aspect(1)
ax.grid()

f1 = lambda x: x
f2 = lambda x: 2/13*(1+3*x)
ax.plot([-.5, 1.5], [f1(-.5), f1(1.5)], color='#ccccff')
ax.plot([-.5, 1.5], [f2(-.5), f2(1.5)], color='#ccccff')

ax.plot([3/4, 3/4], [-.5, 1.5], color='green', linestyle = 'dashed')

x = np.arange(0, 2, 0.1)
y1 = f1(x)
y2 = f2(x)
y = np.maximum(y1, y2)
upper = (lambda x: 1.5)(x)
ax.fill_between(
    x,
    y,
    upper,
    color='#ccccff',
    alpha=0.5,
    interpolate=True,
)

circle = plt.Circle((3/4, 1/2), .03, color='blue', zorder=5)
ax.add_patch(circle)
ax.annotate(r'$(\frac{3}{4}, \frac{1}{2})$', xy=(3/4+.02, 1/2-.12), fontsize=10)

circle = plt.Circle((2/7, 2/7), .03, color='blue', zorder=5)
ax.add_patch(circle)
ax.annotate(r'$(\frac{2}{7}, \frac{2}{7})$', xy=(2/7-.05, 2/7-.15), fontsize=10)

circle = plt.Circle((2/5, 2/5), .03, color='red', zorder=5)
ax.add_patch(circle)
ax.annotate(r'$(\frac{2}{5}, \frac{2}{5})$', xy=(2/5-.2, 2/5+.15), fontsize=10)

ax.set_xlim(0, 1.5)
ax.set_ylim(0, 1.5)

plt.xlabel(r'$\lambda_1$')
plt.ylabel(r'$\lambda_2$')

plt.show()
```

```python
from sklearn.svm import LinearSVC

X = np.array([
    [2, 1],
    [3, -1],
    [0, 1],
])
y = np.array([1, -1, 1])

clf = LinearSVC(C=.1)
result = clf.fit(X, y)
```

```python
import mglearn

mglearn.plots.plot_2d_separator(result, X, fill=False)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)

plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')

plt.show()
```
