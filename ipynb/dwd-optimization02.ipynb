```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_aspect(1)
ax.grid()

circle = plt.Circle((2, 1), .05, color='blue')
ax.add_patch(circle)
ax.annotate(r'$\xi_1 = (2, 1)$', xy=(2+.2, 1), fontsize=10)
ax.annotate(r'$t_1 = +1$', xy=(2+.2, 1-.2), fontsize=10)

circle = plt.Circle((3, -1), .05, color='blue')
ax.add_patch(circle)
ax.annotate(r'$\xi_2 = (3, -1)$', xy=(3-1, -1), fontsize=10)
ax.annotate(r'$t_2 = -1$', xy=(3-1, -1-.2), fontsize=10)

circle = plt.Circle((0, 1), .05, color='blue')
ax.add_patch(circle)
ax.annotate(r'$\xi_3 = (0, 1)$', xy=(0+.2, 1), fontsize=10)
ax.annotate(r'$t_3 = +1$', xy=(0+.2, 1-.2), fontsize=10)

ax.plot([-.5, 3.5], [-1.5, .5], color='red', linewidth='3')

ax.plot([2, 2.5], [1, 0], color='green', linestyle = 'dashed')
ax.plot([2.5, 3], [0, -1], color='purple', linestyle = 'dashed')


ax.set_xlim(-0.5, 3.5)
ax.set_ylim(-1.5, 1.5)

plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')

plt.show()
```
